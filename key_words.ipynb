```python
import nltk
from nltk.tokenize import word_tokenize, RegexpTokenizer
from nltk.tag import pos_tag

nltk.download('averaged_perceptron_tagger')

# Sample vacancy description
vacancy_description = "We are looking for a software engineer with expertise in Python, Java, and experience with machine learning libraries such as TensorFlow and PyTorch."

# Tokenize the text
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(vacancy_description)

# Part-of-speech tagging
pos_tags = pos_tag(tokens)

# Extract key skills (e.g., Nouns and certain Adjectives)
key_skills = [word for word, pos in pos_tags if pos in ['NN', 'NNS', 'JJ']]

print("Key Skills:", key_skills)
```

```python
import spacy

# Load the spaCy English model
# nlp = spacy.load('en_core_web_sm')
nlp = spacy.load('en_core_web_lg')
nlp 

# Sample vacancy description
# vacancy_description = "We are looking for a data scientist proficient in tools like Python, R, SQL, and experience with visualization tools such as Tableau."

vacancy_description = "Опыт работы с рекомендательными системами (необязательно прод) от 1 года. Понимания, что под капотом RecSys алгоритмов. Стек технологий: Опыт работы...', 'responsibility': 'Предлагать и проводить эксперименты для решения возникающих задач. Применять различные модели МО для решения задач, интерпретировать результаты"

# Process the text with spaCy
doc = nlp(vacancy_description)

# Extract tool names using named entity recognition (NER)
# tool_names = [entity.text for entity in doc.ents if entity.label_ == 'PRODUCT']
tool_names = [entity.text for entity in doc.ents if entity.label_ == 'GPE']
# tool_names = [entity.text for entity in doc.ents if entity.label_ == 'ORG']
# take nouns and verbs
tool_names = [entity.text for entity in doc if entity.pos_ in ['PROPN', 'VERB', 'NOUN']]

print("Detected Tool Names:", tool_names)
```
